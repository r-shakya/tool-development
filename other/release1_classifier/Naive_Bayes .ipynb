```python
col=["issue","link","label"]
```

```python
df=pd.read_csv("train_data.csv")
```

```python
df.drop(["Unnamed: 0","Unnamed: 0.1"],axis=1,inplace=True)
df.drop(["1"],axis=1,inplace=True)
```

```python
import nltk

from nltk import word_tokenize
nltk.download('stopwords')



```

```python
from nltk.corpus import  stopwords

stopWords =set(stopwords.words('english'))
```

```python
print(stopWords)
```

```python
import string
```

```python
def dataclean(a):
    remove_punctuation=[char for char in a if char  not in string.punctuation ]
    remove_punctuation="".join(remove_punctuation)
    return [word  for word in remove_punctuation.split() if word.lower() not in stopwords.words('english')]
```

```python
data=df[["issue","label"]]
```

```python
data
```

```python
data.iloc[:,0].apply(dataclean)
```

```python
data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bow_transformer =CountVectorizer(analyzer=dataclean).fit(data["issue"])
bow_transformer.vocabulary_
```

```python
title_bow=bow_transformer.transform(data["issue"])
```

```python
print(title_bow)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
tfidf_transformer=TfidfTransformer().fit(title_bow)
print(tfidf_transformer)
```

```python
title_tfidf=tfidf_transformer.transform(title_bow)
print(title_tfidf)
print(title_tfidf.shape)
```

```python
data
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(title_tfidf, data["label"], test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB
model =MultinomialNB().fit(X_train,y_train)
```

```python
predicted=model.predict(X_test)
print(predicted.shape)
```

```python
print(pd.DataFrame(y_test))
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score
print("Accuracy:\n",accuracy_score(y_test,predicted))
print()
#print(metrics.classification_report(y_test,predicted,target_names=y_test))
print()
print("confusion matrix:\n",metrics.confusion_matrix(y_test,predicted))
```
